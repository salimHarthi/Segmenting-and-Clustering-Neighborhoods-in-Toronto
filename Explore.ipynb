```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import numpy as np
import folium
```

```python
df = pd.read_csv('/content/Toronto.data.csv')
df.head()
```

```python
Toronto_only = df[df['Borough'].str.contains('Toronto')]
Toronto_only.head()
```

```python
map_Toronto = folium.Map(location=[43.659526	, -79.340923], zoom_start=12)
for lat, lng, borough, neighborhood in zip(Toronto_only['Latitude'], Toronto_only['Longitude'], Toronto_only['Borough'], Toronto_only['Neighborhood']):
    label = '{}, {}'.format(neighborhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_Toronto)

map_Toronto
```
